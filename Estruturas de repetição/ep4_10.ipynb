```python
#declarando a variável:

n = int(input())

#configurando método para realizar a construção da pirâmide:

def construindo_piramide(n):
    for linha in range (1,n):
      hifen = (n-linha)*'-'
      um = (linha*2-1)*'1'
      print(hifen+um+hifen)

#chamando a saída do programa

construindo_piramide(n)

```
