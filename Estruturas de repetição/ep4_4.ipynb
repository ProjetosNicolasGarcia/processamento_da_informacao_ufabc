```python
#Definindo as variáveis de entrada:

valor = ''
somador = 0
quantidade = 0


#Método para definir a somatória e a quantidade:

def definindo_termos_media(valor,somador,quantidade):
  while valor != 0:
    valor = int(input())

    if valor != 0:
      somador += valor
      quantidade += 1
  return(valor, somador, quantidade)


valor, somador, quantidade = definindo_termos_media(valor,somador,quantidade)

#Método para calcular a média:

def calculando_media(somador,quantidade):

  if quantidade > 0:
    media = somador / quantidade
    print(f"{media:.2f}")


calculando_media(somador,quantidade)


#Chamando a saída do programa:




```
