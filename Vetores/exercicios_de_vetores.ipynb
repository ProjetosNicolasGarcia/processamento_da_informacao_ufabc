```python
#ep14

# 1. Entrada de dados
numeros_texto = input(" ")
v_str = numeros_texto.split(" ") # Único split permitido

# Convertendo tudo para inteiros no vetor principal para facilitar
v = [0] * 10
for i in range(10):
    v[i] = int(v_str[i])

# 2. Inicialização das variáveis
par = 0
impar = 0
maior = v[0]
menor = v[0]

# 3. Contagem de PARES e ÍMPARES ÚNICOS e definição de maior/menor
for i in range(10):
    valor_atual = v[i]

    # Lógica de maior e menor
    if valor_atual > maior:
        maior = valor_atual
    if valor_atual < menor:
        menor = valor_atual

    # LÓGICA DE VERIFICAÇÃO DE REPETIÇÃO ("Olhar para trás")
    repetido = False
    for j in range(i): # Vai do índice 0 até o índice anterior ao 'i'
        if v[j] == valor_atual:
            repetido = True # Opa, achei um igual lá atrás!

    # Se não for repetido, a gente conta a paridade
    if repetido == False:
        if valor_atual % 2 == 0:
            par += 1
        else:
            impar += 1

# 4. Criando os vetores com o tamanho exato dos números únicos
vpar = [0] * par
vimpar = [0] * impar

# 5. Preenchendo os vetores (usando a mesma lógica de não repetição)
indice_par = 0
indice_impar = 0

for i in range(10):
    valor_atual = v[i]

    # Novamente, checamos se ele já apareceu antes para não guardar duplicatas
    repetido = False
    for j in range(i):
        if v[j] == valor_atual:
            repetido = True

    # Só salva nos novos vetores se for a primeira vez que ele aparece
    if repetido == False:
        if valor_atual % 2 == 0:
            vpar[indice_par] = valor_atual
            indice_par += 1
        else:
            vimpar[indice_impar] = valor_atual
            indice_impar += 1

# Exibindo resultados
print("Pares únicos:", end=" ") # Imprime o texto e NÃO pula linha
for i in range(indice_par):     # Percorre apenas as posições preenchidas
    print(vpar[i], end=" ")     # Imprime o número e um espaço do lado
print() # Esse print vazio serve apenas para pular a linha no final do laço

# Exibindo os Ímpares
print("Ímpares únicos:", end=" ")
for i in range(indice_impar):
    print(vimpar[i], end=" ")
print() # Pula a linha final

# Exibindo Maior e Menor (aqui continua normal)
print("Maior:", maior)
print("Menor:", menor)
```

```python
# Declarando as variáveis de entrada
numeros_texto = input()
v_str = numeros_texto.split(" ")

numeros_texto2 = input()
v_str2 = numeros_texto2.split(" ")

# Configurando vetor 1
v = [0] * 5
for i in range(5):
    v[i] = int(v_str[i])

# Configurando vetor 2
v2 = [0] * 5
for i in range(5):
    v2[i] = int(v_str2[i])

# Configurando vetor de intersecção
# O tamanho máximo da intersecção é 5 (caso todos os elementos sejam iguais e únicos)
vi = [0] * 5
indice_vi = 0

for i in range(5):
    for j in range(5):
        if v[i] == v2[j]:
            # Elemento encontrado em v2. Verificar se já está no vetor de intersecção (vi)
            ja_existe = False
            for k in range(indice_vi):
                if vi[k] == v[i]:
                    ja_existe = True
                    break # Interrompe a busca no vetor resultante, pois já encontrou

            # Se não existe no vetor resultante, adiciona
            if not ja_existe:
                vi[indice_vi] = v[i]
                indice_vi += 1

            break # Interrompe a busca em v2 para este elemento de v

# Montando a saída baseada apenas no número de elementos reais da intersecção
formatado = ""
for i in range(indice_vi):
    formatado += str(vi[i]) + " "

print("O vetor interseccao eh", formatado)
```

```python
#ep22

#imports:

import math

#variáveis de entrada:

coordenadas_texto = input()
c_str = coordenadas_texto.split(" ")

n = int(input())

numeros_texto = input()
v_str = numeros_texto.split(" ")





#Configurando vetor:

c = [0] * 2

for i in range(2):
    c[i] = int(c_str[i])

v = [0] * n

for i in range(n):
    v[i] = int(v_str[i])

#Exibindo vetor

formatado = ""

for i in range(n):
  formatado += str(v[i]) + " "


print(formatado)

#Configurando a distância de x:

# 1. Alocação correta FORA do laço (n // 2 garante um número inteiro)
tamanho_v2 = n // 2
v2 = [0] * tamanho_v2

# 2. Laço de cálculo
for i in range(0, n, 2):
    d = math.sqrt(((v[i] - c[0]) ** 2) + ((v[i+1] - c[1]) ** 2))

    # 3. Mapeamento de índice: i // 2 converte (0, 2, 4) em (0, 1, 2)
    indice_v2 = i // 2
    v2[indice_v2] = d

formatado2 = ""

formatado2 = ""

for i in range(tamanho_v2):
    # A formatação :.2f atua diretamente sobre o float v2[i] antes de virar texto
    formatado2 += f"{v2[i]:.2f} "

# Imprime a string que já foi construída com os números formatados
print(formatado2)

```

```python
#ep26

# 1. Lendo os centros iniciais
linha = input().split()
n = len(linha)
centros = [0.0] * n
for i in range(n):
    centros[i] = float(linha[i])

# 2. Lendo as cores (pontos a serem agrupados)
linha = input().split()
ncor = len(linha)
cores = [0.0] * ncor
for i in range(ncor):
    cores[i] = float(linha[i])

# Vetor para armazenar a qual centro cada cor pertence
proximo = [0] * ncor

# 3. Associando cada cor ao centro mais próximo
for j in range(ncor):
    distancia_minima = float('inf') # Substitui o limite arbitrário de 120
    indice = 0
    for i in range(n):
        distancia_atual = abs(cores[j] - centros[i])
        if distancia_minima > distancia_atual:
            distancia_minima = distancia_atual
            indice = i
    proximo[j] = indice

# 4. Calculando as médias para encontrar os novos centros
novos_centros = [0.0] * n
soma_centros = [0.0] * n
conta_centros = [0] * n

for i in range(n):
    for j in range(ncor):
        # Verifica corretamente se a cor 'j' pertence ao centro 'i'
        if proximo[j] == i:
            soma_centros[i] += cores[j]
            conta_centros[i] += 1

    # Calcula a média se o centro recebeu algum ponto, senão mantém o original
    if conta_centros[i] > 0:
        novos_centros[i] = soma_centros[i] / conta_centros[i]
    else:
        novos_centros[i] = centros[i]

# 5. Exibindo o resultado no formato exato exigido pelo corretor
for i in range(n):
    # O índice i começa em 0, então somamos 1 para imprimir "Centro 1, Centro 2..."
    print(f"Centro {i+1} ({centros[i]:.2f}): {novos_centros[i]:.2f}")
```
