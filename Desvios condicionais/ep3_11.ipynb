```python
# declarando as variáveis de entrada:

teta = float(input())
distancia = int(input())

#Método para realizar a análise dos ângulos fornecidos:

def analisando_angulos(teta):
  if ((teta == 0) or (teta < 0) or (teta > 90)):
    print("ERRO: Angulo zero indica medicao no equador ou Sol no zenite")
    booleano = 0
  elif (teta < 1):
    print("AVISO: Angulo muito pequeno (< 1 grau) - medicao pode ser imprecisa")
    booleano = 1
  elif (teta > 89):
    print("AVISO: Angulo muito grande (> 90 graus) - medicao pode ser imprecisa")
    booleano = 1
  else:
    booleano = 1
  return(booleano)


#Método para realizar a análise da distância:

def analisando_distancia(distancia):
  if  ((distancia < 500) and (distancia > 0)) or (distancia > 2000):
    print("AVISO: Distancia fora da faixa esperada (500 a 2000 km)")
  else:
    pass

#Método para calcular a circunferência da Terra:

def calculando_circunferencia(teta, distancia):
  c = (360 / teta) * distancia
  print(f"Circunferencia da Terra: {c:.2f} km")
  return(c)

#Método para calcular o erro:

def calculando_erro(circunferencia):
  e = (abs(circunferencia - 40075) / 40075) * 100
  print(f"Erro: {e:.2f} %") # Formatando para 2 casas decimais
  return(e)

#Método para comparar erro:

def comparando_erro(erro):
  if erro < 5:
    print("Resultado: EXCELENTE")
  elif (erro >= 5) and (erro < 15):
    print("Resultado: BOM")
  elif (erro >= 15) and (erro < 30):
    print("Resultado: ACEITAVEL")
  elif (erro >= 30) and (erro < 50):
    print("Resultado: RUIM")
  else:
    print("Resultado: MEDICAO INVALIDA")



#Montando a saída do exercício

b = analisando_angulos(teta)

def comparando_cenarios(b):
  if b == 1:
    analisando_distancia(distancia)
    circunferencia = calculando_circunferencia(teta, distancia)
    erro = calculando_erro(circunferencia)
    comparando_erro(erro)
  else:
    pass

comparando_cenarios(b)

```
